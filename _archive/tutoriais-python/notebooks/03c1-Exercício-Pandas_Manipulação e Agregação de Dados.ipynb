Inicie um DataFrame a partir do arquivo `2017-02-01_156_-_Base_de_Dados_sample.csv`. 

**Dica:** Caso o arquivo não seja separado por vírgulas, o Pandas não conseguirá reconhecer os campos adequadamente. Você precisará fornecer o parâmetro `sep=';'` para indicar o separador correto (neste caso, o `;`).

**Dica:** Aplicativos modernos tendem a armazenar arquivos no formato UTF-8, mas é comum encontrar arquivos codificados em outros formatos. Caso você tenha problemas para ler o arquivo, utilize o parâmetro `encoding='latin-1'` para selecionar o encoding correto (*latin-1* neste exemplo).

```python
# Resposta:
import pandas as pd
# lê o arquivo CSV
df = pd.read_csv('../data/2017-02-01_156_-_Base_de_Dados_sample.csv', sep=';', encoding='latin-1')
```

Crie um novo DataFrame chamado `df_elogios` contendo apenas as linhas cujos valores na coluna `TIPO` são iguais a 'ELOGIO'. Este DataFrame deve ter as colunas TIPO, ASSUNTO, e BAIRRO_CIDADAO.

```python
# Resposta:
df_elogios = df[df['TIPO'] == 'ELOGIO'][['TIPO', 'ASSUNTO', 'BAIRRO_CIDADAO']]
df_elogios
```

Quais são os 5 bairros com mais elogios? Agrupe pelo `BAIRRO_CIDADAO` e ordene o resultado para responder.

```python
# Resposta:
top_bairros = df_elogios.groupby('BAIRRO_CIDADAO').size().sort_values(ascending=False).head(5)
print(top_bairros)
```

Crie dois novos DataFrames, um contendo os registros de pessoas do sexo masculino e outra do sexo feminino. Use estes DataFrames para responder:

Qual é a proporção de reclamações entre homens e mulheres? (total de reclamações de homens dividido pelo total de reclamações de mulheres)

```python
# Resposta:
df_homens = df[df['SEXO'] == 'M']
df_mulheres = df[df['SEXO'] == 'F']

# Filtra apenas reclamações
reclamacoes_homens = df_homens[df_homens['TIPO'] == 'RECLAMAÇÃO']
reclamacoes_mulheres = df_mulheres[df_mulheres['TIPO'] == 'RECLAMAÇÃO']

# Calcula a proporção
proporcao = len(reclamacoes_homens) / len(reclamacoes_mulheres)
print("Proporção de reclamações entre homens e mulheres:", proporcao)
```

Concatene os dois DataFrames criados acima para criar um DataFrame chamado `df_todos`. Verifique se este DataFrame tem o mesmo número de linhas do DataFrame lido do CSV.

```python
# Resposta:
df_todos = pd.concat([df_homens, df_mulheres])
print("Linhas em df_todos:", len(df_todos))
print("Linhas no DataFrame original:", len(df))
print("Quantidade igual?", len(df_todos) == len(df))
```

Crie um novo DataFrame chamado `df_sexo` contendo a contagem de registros para cada sexo. A coluna com a contagem deve se chamar `contagem`.

```python
# Resposta:
df_sexo = df['SEXO'].value_counts().rename_axis('SEXO').reset_index(name='contagem')
df_sexo
```

Crie um novo DataFrame chamado `df_sexo_tipo` contendo a contagem de registros para cada sexo e tipo. A coluna com a contagem deve se chamar `total`.

```python
# Resposta:
df_sexo_tipo = df.groupby(['SEXO', 'TIPO']).size().reset_index(name='total')
df_sexo_tipo
```

Faça uma junção (join ou merge) dos DataFrames `df_sexo, df_sexo_tipo` para criar um novo DataFrame chamado `df_juncao` contendo as contagens por sexo e também por sexo/tipo.

```python
# Resposta:
df_juncao = pd.merge(df_sexo_tipo, df_sexo, on='SEXO')
df_juncao
```

Crie uma nova coluna no DataFrame `df_juncao` chamada `proporcao` contendo a porcentagem de homens e mulheres que fizeram cada tipo de registro. Responda: qual sexo faz mais elogios? Qual faz mais solicitações? Qual faz mais reclamações?

```python
# Resposta:
df_juncao['proporcao'] = df_juncao['total'] / df_juncao['contagem'] * 100
display(df_juncao)

# Para responder:
elogios = df_juncao[df_juncao['TIPO'] == 'ELOGIO']
solicitacoes = df_juncao[df_juncao['TIPO'] == 'SOLICITAÇÃO']
reclamacoes = df_juncao[df_juncao['TIPO'] == 'RECLAMAÇÃO']

print("Sexo com mais elogios:", elogios.loc[elogios['total'].idxmax(), 'SEXO'])
print("Sexo com mais solicitações:", solicitacoes.loc[solicitacoes['total'].idxmax(), 'SEXO'])
print("Sexo com mais reclamações:", reclamacoes.loc[reclamacoes['total'].idxmax(), 'SEXO'])
```
